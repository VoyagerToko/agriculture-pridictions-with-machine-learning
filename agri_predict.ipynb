```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/RS_Session_260_AU_1830_1.csv")
df.head()
```

```python
df.tail()
```

```python
remove = "Annual Average Retail Prices"
df['Food Commodities'] = df['Food Commodities'].str.replace(remove,"avg_price", regex=False)
df.head()
```

```python
df = df.drop(['Units'], axis = 1)
```

```python
df
```

```python
df.drop_duplicates()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Sl. No.'], axis=1)
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
df_melted = df.melt(id_vars=["Food Commodities", "State/UT"], var_name="Year", value_name="Price")
df_melted["Year"] = pd.to_datetime(df_melted["Year"], format='%Y')

print(df_melted)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for state in df["State/UT"]:
    subset = df_melted[df_melted["State/UT"] == state]
    plt.plot(subset["Year"], subset["Price"], marker='o', label=state)

plt.title("Price Trend (2018-2023)")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram for a single column
df['2018'].hist(bins=20)
plt.title('Histogram of 2018 Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
df[['2018', '2019', '2020', '2021', '2022', '2023']].boxplot()
plt.title('Box Plot of Annual Prices')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter plot for two columns
plt.scatter(df['2018'], df['2019'])
plt.title('Scatter Plot of 2018 vs 2019 Prices')
plt.xlabel('2018 Prices')
plt.ylabel('2019 Prices')
plt.show()

```

```python
# Scatter plot for two columns
plt.scatter(df['2018'], df['2020'])
plt.title('Scatter Plot of 2018 vs 2020 Prices')
plt.xlabel('2018 Prices')
plt.ylabel('2020 Prices')
plt.show()
```

```python
# Scatter plot for two columns
plt.scatter(df['2020'], df['2023'])
plt.title('Scatter Plot of 2020 vs 2023 Prices')
plt.xlabel('2020 Prices')
plt.ylabel('2023 Prices')
plt.show()
```

```python
# Line plot showing the trend over years for a particular state/UT
df[df['State/UT'] == 'Maharashtra'][['2018', '2019', '2020', '2021', '2022', '2023']].T.plot()
plt.title('Annual Price Trend for Maharashtra')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend(['Maharashtra'])
plt.show()
```

```python
import seaborn as sns

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[['2018', '2019', '2020', '2021', '2022', '2023']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Bar plot for average price in 2023 by state
df.groupby('State/UT')['2023'].mean().sort_values().plot(kind='bar', figsize=(12, 6))
plt.title('Average 2023 Prices by State/UT')
plt.ylabel('Average Price')
plt.show()
```

```python
# Violin plot to show distribution of 2023 prices across different commodities
plt.figure(figsize=(12, 6))
sns.violinplot(x='Food Commodities', y='2023', data=df)
plt.xticks(rotation=90)
plt.title('Violin Plot of 2023 Prices by Food Commodities')
plt.show()
```
